### Imports

```python
import requests
from bs4 import BeautifulSoup
```

### Imprimindo o conte√∫do do site

```python
response = requests.get('https://g1.globo.com/')

print(response.content)
```

### Transformando o conteudo do site em uma variavel BeautifulSoup

```python
content = response.content

site = BeautifulSoup(content , 'html.parser')

type(site)
```

#### Organizando melhor a visualiza√ß√£o do conteudo

```python
site.prettify()
```

### Extraindo a noticia

```python
noticia = site.find('div', attrs={'class': 'feed-post bstn-item-shape type-materia'})
noticia
```

### Extraindo o t√≠tulo

```python
titulo = noticia.find('a', attrs={'class':'feed-post-link'})
titulo.text
```

### Extraindo o subt√≠tulo

```python
subtitulo = noticia.find('div', attrs={'class': 'feed-post-body-title'})
subtitulo.text
```
